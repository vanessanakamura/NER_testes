```python
num_labels = {
    "0": "O", "1": "B-ORGANIZACAO", "2": "I-ORGANIZACAO", "3": "B-PESSOA",
    "4": "I-PESSOA", "5": "B-TEMPO", "6": "I-TEMPO", "7": "B-LOCAL",
    "8": "I-LOCAL", "9": "B-LEGISLACAO", "10": "I-LEGISLACAO",
    "11": "B-JURISPRUDENCIA", "12": "I-JURISPRUDENCIA"
}
```

```python
# Converter chaves de string para inteiros no dicionário
num_labels = {int(k): v for k, v in num_labels.items()}

# Remover prefixos 'B-' e 'I-' das etiquetas para obter apenas os tipos de entidade
tag_map = {k: v.replace('B-', '').replace('I-', '') for k, v in num_labels.items()}
```

```python
relevant_entities = [
    "ORGANIZACAO", "PESSOA", "TEMPO", "LOCAL", "LEGISLACAO", "JURISPRUDENCIA"
]
```

```python
filtered_tag_map = {k: v for k, v in tag_map.items() if v in relevant_entities}
filtered_tag_map
```

# Dataset Lener FineTune

```python
def group_entities(tokens, tags):
    """
    Agrupa entidades com base nos tokens e tags fornecidos.
    - tokens: lista de palavras/token do texto.
    - tags: lista de tags correspondentes a cada token que indicam o tipo de entidade.
    Retorna um dicionário onde cada chave é um tipo de entidade e o valor é uma lista de entidades detectadas desse tipo.
    """
    grouped_entities = {entity_type: [] for entity_type in filtered_tag_map.values()}
    current_entity = None
    current_type = None

    for token, tag in zip(tokens, tags):
        tag_str = num_labels[tag]
        entity_type = filtered_tag_map.get(tag)

        if entity_type:
            if tag_str.startswith('B-'):
                if current_entity is not None:
                    grouped_entities[current_type].append(current_entity)
                current_entity = token
                current_type = entity_type
            elif tag_str.startswith('I-') and current_type == entity_type:
                current_entity += ' ' + token
            else:
                if current_entity is not None:
                    grouped_entities[current_type].append(current_entity)
                current_entity = None
                current_type = None

    if current_entity is not None:
        grouped_entities[current_type].append(current_entity)

    return grouped_entities


def split_tokens(tokens, ner_tags, min_length=50, max_length=100):
    """
    Divide uma lista de tokens e suas respectivas tags em segmentos.
    - tokens: lista de palavras/token do texto.
    - ner_tags: lista de tags de entidade correspondentes aos tokens.
    - min_length: comprimento mínimo de um segmento.
    - max_length: comprimento máximo de um segmento.
    Retorna uma lista de tuplas, onde cada tupla contém os tokens e as tags de um segmento.
    """
    segments = []
    start = 0

    while start < len(tokens):
        end = min(start + max_length, len(tokens))
        while end > start + min_length and end > start and (len(tokens[end - 1]) == 0 or tokens[end - 1][-1] != '.'):
            end -= 1
        if end <= start + min_length:
            end = min(start + max_length, len(tokens))

        segment_tokens = tokens[start:end]
        segment_tags = ner_tags[start:end]
        segments.append((segment_tokens, segment_tags))

        start = end

    return segments


instruction = """
Extract relevant information from the provided text based on the following predefined entities. Each entity corresponds to a specific description that defines the nature of the information to be extracted. Organize the extracted data in a structured JSON format, ensuring that each field contains only relevant information according to its description. If a specific entity is not found in the text, omit it from the output. The expected entities and their descriptions are:

Relevant Entities:
nome: The specific name of the fund, which uniquely identifies it. The name must be longer than three words.
cnpj_fundo: The official CNPJ number assigned to the fund for legal and administrative identification.
condominio_aberto: Indicates whether the fund operates as an open-ended fund, offering continuous buy/sell opportunities.
composicao_n: Describes the internal composition or structure of the fund, if specified.
tipo_investidor: Identifies the type of investor the fund is intended for (e.g., general public, qualified investors).
administrador: The institution responsible for the administrative management of the fund.
Gestor: The entity or company responsible for managing the assets and investments of the fund.
Custodiante: The custodian responsible for holding and safeguarding the fund’s financial assets.
investimento_ext: Information about any external investments made by the fund, if mentioned.
tx_administracao: The annual management fee percentage charged for fund administration.
"""

def save_valid_samples_to_json(dataframe, json_filename):
    """
    Filtra amostras com entidades válidas e salva em um arquivo JSON.

    Parameters:
    dataframe (pd.DataFrame): DataFrame contendo as amostras.
    json_filename (str): Nome do arquivo JSON onde as amostras serão salvas.
    """
    valid_samples = []

    # Itera sobre as amostras no DataFrame
    for idx, sample in dataframe.iterrows():
            segments = split_tokens(sample['tokens'], sample['ner_tags'], min_length=100, max_length=200)

            for segment_tokens, segment_ner_tags in segments:
                grouped_entities = group_entities(segment_tokens, segment_ner_tags)

                # Excluir chaves com listas vazias
                grouped_entities = {k: v for k, v in grouped_entities.items() if v}

                if grouped_entities:  # Verifica se grouped_entities não está vazio
                    text = ' '.join(segment_tokens)

                    # Adiciona o dicionário no formato desejado à lista
                    valid_samples.append({
                        "instruction": instruction,
                        "input": text,
                        "output": grouped_entities
                    })

    # Salva os resultados no formato JSON
    with open(json_filename, "w", encoding="utf-8") as f:
        json.dump(valid_samples, f, ensure_ascii=False, indent=4)

    print(f"JSON salvo com sucesso em {json_filename}!")
```

```python
save_valid_samples_to_json(train_df, "train_finetune_lenerbr.json")
save_valid_samples_to_json(test_df, "test_finetune_lenerbr.json")
```
